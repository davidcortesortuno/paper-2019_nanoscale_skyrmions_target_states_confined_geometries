# 2D simulation

There are two mesh options for the 2D simulations, according to the `simulation` argument: `'2D_square'` and `experimental_sample`. The former is a square shaped finite layer with a default dimension of 50 x 50 spins. The latter uses a mesh based on an experimental sample of PdFe islands.

The magnetic field applied to the sample can be specified for every component. For a start, it is recommended to only update the component in the $z$ direction

The following are the main steps to start one of these interactive simulations. We remove the precession from the LLG equation to speed up the relaxation process.

```python
i_sim = interactive_simulation(simulation='experimental_sample',                          
                               ku=0.41,  # meV,
                               mu_s=3,   # mu_B
                               J=5.88    # meV
                               )
i_sim.generate_simulation(do_precession=False,
                          load_mu_s='mu_s/Pd_Fe_Ir111_20160118_CrTip_015_B-4T_003_rotated_large_bw_mus3-muB.npy'
                          )
i_sim.generate_m_field()

i_sim.generate_widgets()
```
